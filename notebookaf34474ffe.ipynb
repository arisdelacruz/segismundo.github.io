```python
import pandas as pd
biofuels = pd.read_csv('/kaggle/input/biofuel-consumption/biofuels_consumption.csv')
biofuels.info()
```

```python
biofuels = biofuels.replace(r"^ +| +$", r"", regex=True)
```

```python
biofuels_ten = biofuels.loc[biofuels['Country'].isin(['Belgium',
                                                       'Canada',
                                                       'France',
                                                       'Germany',
                                                       'Italy',
                                                       'Japan',
                                                       'Netherlands',
                                                       'Sweden',
                                                       'Switzerland',
                                                       'United Kingdom',
                                                       'United States'])]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
chart_ten = sns.barplot(x="2020", y="Country", data=biofuels_ten)
chart_ten.set_xlabel('Year: 2020')
chart_ten.set_ylabel('Name of Country')
chart_ten.set_title('Biofuel Consumption of G10 Countries\nduring the Pandemic (Year 2020)',
             fontdict={'size': 12,'color': 'blue'})
```
